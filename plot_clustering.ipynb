### Plot de agrupamento (*clustering*) com 3 ou mais atributos

Neste exemplo é mostrado como plotar agrupamentos quando se tem mais de um atributo.

```python
import numpy as np
from urllib2 import urlopen
from sklearn.datasets import load_digits

digits = load_digits()
print digits.data
print digits.data.shape
```

Aqui temos o *dataset* `digits` carregada com 64 atributos (pixels representando cada imagem). Porém não podemos mostrar estes dados com 64 dimensões graficamente.

O *Principal Component Analysis* (PCA) é um método não supervisionado rápido e flexível para a redução da dimensionalidade dos dados. Podemos utilizar este método tanto para demonstrar os dados no gráfico quanto para aplicar os algoritmos de agrupamento com os dados gerados por ele.

O exemplo abaixo mostra a redução das **sessenta e quatro** dimensões do *dataset* para **duas** dimensões:

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
projected = pca.fit_transform(digits.data)
print projected
print 'Dimensão:', projected.shape
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(projected[:, 0], projected[:, 1], alpha=0.5)
```

O `scatter plot` está mostrando todas as imagens, porém, não é possível ter ideia de nada sobre os dados pois todos eles está em uma escala $\alpha = 0.5$ de azul.

Vamos supor que os próprios rótulos no *dataset* `digits` são os nosso "clusters" e cada *cluster* possui uma diferente cor:

```python
plt.scatter(projected[:, 0], projected[:, 1], c=digits.target, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('nipy_spectral_r', 10))
plt.xlabel('Componente 1')
plt.ylabel('Componente 2')
plt.colorbar()
```

Para plotar o resultado dos agrupamentos para o trabalho, utilizem a quantidade final de agrupamento como rótulo, como foi feito no exemplo acima.
